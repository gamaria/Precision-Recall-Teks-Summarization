```python
import csv


a = []
with open("data_pakar.csv", "r") as f:
     pakar = [x for x in csv.reader(f)]
     for u in range(0, len(pakar)):
         for t in pakar[u]:
             a.append(t)
b = []
with open("data_sistem.csv", "r") as d:
     sistem = [j for j in csv.reader(d)]
     for q in range(0, len(sistem)):
         for w in sistem[q]:
             b.append(w)

jumlahFN = []
x = []
FN = []
for em in range(0, len(pakar)):
    j = []
    for im in pakar[em]:
        if im in sistem[em]:
           pass
        else:
            jumlahFN.append(im)
            x.append(im)
            j.append(im)
    FN.append(j)


jumlahFP = []
FP = []
print(len(x))
for e in range(0, len(sistem)):
    r = []
    for f in sistem[e]:
        if f in pakar[e]:
           pass
        else:
            jumlahFP.append(f)
            r.append(f)
    FP.append(r)
#print('FP', FP)
#print('FN',FN)
total_TP = len(a)
print('total data TP pakar',total_TP)
```

```python
jum_FP =len(jumlahFP)
jum_FN = len(jumlahFN)
print ("total data FN =", jum_FP)
print ("total data FN =", jum_FN)
```

```python
#Precession Recall
pre= total_TP/(total_TP+jum_FP)
rec = total_TP/(total_TP+jum_FN)
f_score = (2* pre*rec)/(pre+rec)

print ('precision=', float(pre))
print ('recall=', rec)
print ('f-measure', f_score)
```
